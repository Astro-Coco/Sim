```python
data['Thrust avg'] = data['Thrust'].rolling(window = 40, min_periods=0).mean()
```

```python
plt.scatter(data['time'], data['Thrust'], s = 1, label = 'Poussée')
plt.scatter(data['time'], data['Thrust avg'], s = 1, label = 'Poussée Moyenne')
plt.scatter(data['Time      (s)'], data['Thrust (sim)'], s = 1, label = 'Poussée Simulée')

plt.legend()

plt.minorticks_on()
plt.grid(which='both', linestyle='-', linewidth='0.5', alpha=0.5)

plt.xlabel('Temps (s)')
plt.ylabel('Poussée (N)')
plt.show()
```

```python
data.columns
```

```python
plt.scatter(data['time'], data['Pressure Tank'], s = 1, label = 'Pression Bonbonne')
plt.scatter(data['Time      (s)'], data['Pressure Tank (sim)'], s = 1, label = 'Pression Bonbonne (sim)')

plt.scatter(data['time'], data['Combustion chamber pressure'], s = 1, label = 'Pression Chambre de Combustion')
plt.scatter(data['Time      (s)'], data['Pressure combustion chamber (sim)'], s = 1, label = 'Pression Chambre de Combustion (sim)')

plt.legend()

plt.minorticks_on()
plt.grid(which='both', linestyle='-', linewidth='0.5', alpha=0.5)

plt.xlabel('Temps (s)')
plt.ylabel('Pression (psi)')
plt.show()
```

```python
plt.scatter(data['time'], data['Thrust avg_integral'], s = 1, label = 'Impulsion')
plt.scatter(data['Time      (s)'], data['Impulse (sim)'], s = 1, label = 'Impulsion simulée')

plt.text(x = 3.9, y = 12000, s = f'''Impulsion totale sim : {max(data['Impulse (sim)']):.2f} Ns''')
plt.text(x = 3.9, y = 10000, s = f'''     Impulsion totale: {max(data['Thrust avg_integral']):.2f} Ns''')

plt.legend()

plt.minorticks_on()
plt.grid(which='both', linestyle='-', linewidth='0.5', alpha=0.5)

plt.xlabel('Temps (s)')
plt.ylabel('Impulsion (Ns)')
plt.show()
```
